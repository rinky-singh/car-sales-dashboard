```python
# IMPORTING ESSENTIAL LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# LOADING AND VIEWING DATASET

air = pd.read_csv('/content/drive/MyDrive/python/Airbnb_Open_Data.csv')
air
```

```python
# SIZE OF DATASET

print('Rows, Columns: ', air.shape)
print('Number of Elements:, ', air.size)
```

```python
# GENERAL INFORMATION ABOUT THE DATASET

air.info()
```

```python
# CHECKING DATA-TYPES OF COLUMNS

air.dtypes
```

```python
# FIRST 10 ROWS OF THE DATASET

air.head(10)
```

```python
# LAST 10 ROWS OF THE DATASET

air.tail(10)
```

```python
# STATISTICS ABOUT NUMERIC COLUMNS IN THE DATASET

air.describe()
```

```python
# CHECKING UNIQUE VALUES IN CERTAIN COLUMNS

print('Number of unique values in Neighbourhood Group column: ', air['neighbourhood group'].nunique())
print('Number of unique values in Neighbourhood column: ',air['neighbourhood'].nunique())
print('Number of unique values in Room Type column: ',air['room type'].nunique())
print('Number of unique values in Service Fee column: ', air['service fee'].nunique())
```

```python
# CONVERTING PRICE COLUMN FROM STRING TYPE TO FLOAT TYPE

air['price'] = air['price'].str.replace('$', '')
air['price'] = air['price'].str.replace(',', '')
air['price'] = air['price'].astype(float)
```

```python
# RE-CHECKING STATISTICS OF NUMERIC COLUMNS

air.describe()
```

**VISUALIATION**

```python
# PRICE DISTRIBUTION

sns.histplot(air['price'], kde=True)
```

```python
# ROOM-TYPE DISTRIBUTION

sns.countplot(y='room type', data = air)
```

```python
# PRICE V/S ROOM-TYPE

sns.boxplot(x='room type', y='price', data = air)
```

```python
# REVIEWS PER MONTH DISTRIBUTION

sns.histplot(air['reviews per month'], kde=True)
```

```python
# AVAILABILITY BY NEIGHBOURHOOD GROUP

sns.boxplot(x='neighbourhood group', y='availability 365', data = air)
```

```python
# CORRELATION HEATMAP

corr = air.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True)
```

```python
# PRICE V/S NUMBER OF REVIEWS

sns.scatterplot(x='number of reviews', y='price', data = air)
```

```python
# PRICE COMPARISON BY NEIGHBOURHOOD GROUP

plt.figure(figsize=(10, 5))
sns.boxplot(x='neighbourhood group', y='price', data = air)
plt.title('Price Comparison by Neighborhood Group')
plt.xticks(rotation=45)
plt.show()
```

```python
# REVIEWS BY NEIGHBOURHOOD GROUP

plt.figure(figsize=(10, 5))
neighborhood_reviews = air.groupby('neighbourhood group')['number of reviews'].sum().reset_index()
sns.barplot(x='neighbourhood group', y='number of reviews', data=neighborhood_reviews)
plt.title('Reviews by Neighborhood Group')
plt.xticks(rotation=45)
plt.show()
```
